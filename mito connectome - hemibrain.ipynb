```python
fs = 1.2 # hz
cutoff = 0.1 # hz - cutoff frequency for butterworth filter

x_start = 100
dt = 400  # datapts
timevec = np.arange(dt) / fs  # sec
dff_thresh = 0.15
def exp_func(k, x, y):
    return np.sum( (y - np.exp(-x/k))**2 )

indicator_roi_fracActive_T_std = []

for plt_all in [True, False]:
    ycolors = {'Calcium': '#FF7F3E', 'ATP': '#003285', 'Pyruvate': '#008B8B'}
    for indicator, roi_responses in zip(['Calcium', 'ATP', 'Pyruvate'], [ca_responses, atp_responses, pyruvate_responses]):
        if indicator == 'Calcium':
            dy = 0.5
        elif indicator == 'ATP':
            dy = 0.8
        else:
            dy = 0.2
        roi_fracActive_T_std = []
        fig, ax = plt.subplots(figsize=(7,10)) if plt_all else plt.subplots(figsize=(6,2.5))
        fontsize = 8 if plt_all else 18
        y_roi = 0
        for ii, roi in enumerate(np.array(rois)[bool_rois]):
            this_frac_active, this_std = [], []
            this_corrs = []
            for roi_response in roi_responses[ii]:
                roi_response_filtered = roi_response[x_start : x_start+dt]
                reg = sm.OLS(roi_response_filtered, sm.add_constant(timevec) ).fit()
                pred_ys = reg.predict( sm.add_constant(timevec) )
                dff = (roi_response_filtered - pred_ys) / pred_ys
                this_corrs.append( np.correlate(dff, dff, 'full')[dt-1:] )
                this_frac_active.append( np.mean(dff > dff_thresh) )
                this_std.append( np.std(dff) )
            this_corrs = np.array(this_corrs) / np.array(this_corrs)[:,[0]]
            res = minimize( exp_func, (5), args = (timevec, this_corrs), method = 'Nelder-Mead')
            roi_fracActive_T_std.append( [np.mean(this_frac_active), res.x[0], np.mean(this_std)] )
            
            taus = []
            for this_corr in this_corrs:
                res = minimize( exp_func, (5), args = (timevec, this_corr[np.newaxis,:]), method = 'Nelder-Mead')
                taus.append(res.x[0])

            if plt_all or (roi in ['PLP(R)', 'BU(L)', 'CRE(R)', 'SLP(R)']):
                i_neuropil_class = np.where( np.array([ roi in neuropils for neuropils in config.all_neuropils]) )[0][0]
                color = config.neuropil_colors[i_neuropil_class]

                y_roi -= dy
                roi_response = roi_responses[ii][np.argmax(this_std)]
                roi_response_filtered = roi_response[x_start : x_start+dt]
                reg = sm.OLS(roi_response_filtered, sm.add_constant(timevec) ).fit()
                pred_ys = reg.predict( sm.add_constant(timevec) )
                dff = (roi_response_filtered - pred_ys) / pred_ys
                if False:#plt_all == False:
                    ax.plot(timevec[:150], dff[:150] + y_roi, linewidth = 0.5, color =  color)
                else:
                    ax.plot(timevec, dff + y_roi, linewidth = 0.5, color =  color)
                ax.text(-5, y_roi, roi, verticalalignment = 'center', horizontalalignment = 'right', fontsize=fontsize)
        y_roi = y_roi - 2*dy if plt_all else -dy*2.5#y_roi - dy
        dx = 0 if plt_all else timevec[-1]*1.2
        ax.plot([0+dx,60+dx], [y_roi, y_roi], linewidth = 0.5, color = 'k')
        
        if indicator == 'Pyruvate':
            ax.text(30+dx, y_roi-0.04, '1 min', verticalalignment = 'top', horizontalalignment = 'center', fontsize=fontsize)
            ax.plot([0+dx,0+dx], [y_roi, y_roi+0.1], linewidth = 0.5, color = 'k')
            ax.text(0+dx-5, y_roi+0.05, '10%', verticalalignment = 'center', horizontalalignment = 'right', fontsize=fontsize)
        else:
            ax.text(30+dx, y_roi-0.1, '1 min', verticalalignment = 'top', horizontalalignment = 'center', fontsize=fontsize)
            ax.plot([0+dx,0+dx], [y_roi, y_roi+0.3], linewidth = 0.5, color = 'k')
            ax.text(0+dx-5, y_roi+0.15, '30%', verticalalignment = 'center', horizontalalignment = 'right', fontsize=fontsize)
        roi_fracActive_T_std = np.array(roi_fracActive_T_std)
        ax.axis('off')
        ax.set_title(f'Example {indicator} Responses', fontsize=fontsize, color = ycolors[indicator])
        file_end = '_all' if plt_all else ''
        plt.savefig(home_dir + f'/saved_figures/nonautonomous/Hemibrain_MitoConnectome/example_{indicator}_traces{file_end}.svg', transparent=True, bbox_inches='tight')
        plt.show()
        
        indicator_roi_fracActive_T_std.append( roi_fracActive_T_std )
```
