```python
def save_as_swc(cellid):    
    filename = "smo_data/cell_morphologies/{:d}.swc".format(cellid)
    try:
        nrn, labels = skel.read_skel_and_labels(cellid)
        print("Exporting to {:s}".format(filename))
        nrn.export_to_swc("smo_data/cell_morphologies/{:d}.swc".format(cellid), node_labels= labels, xyz_scaling=1)
    except Exception as e:
        print("Problem with cell {:s}".format(cellid))
        print(e)

def export_to_swc(cell_file):
    df = pandas.read_csv(cell_file, index_col=None, header=None,names=["cell_id"])
    df.apply(lambda sr: save_as_swc(sr["cell_id"]), axis=1)
    
```

```python
export_to_swc("smo_data/inh_cell.csv")
```
